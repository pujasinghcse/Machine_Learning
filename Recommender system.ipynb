```python
import pandas as pd
```

```python
set1 = pd.read_csv('/home/puja/Downloads/movie_metadata.csv')
print set1.shape
set1.head(3)
```

```python
# v is the number of votes for the movie
# m is the minimum votes required to be listed in the chart
# R is the rating for the movie
# C is the mean vote across the whole report

C = set1['imdb_score'].mean()
print C
```

```python
m = set1['num_voted_users'].quantile(0.90)
print m
```

```python
# Filter out all qualified movies into a new DataFrame
q_movies = set1.copy().loc[set1['num_voted_users']>=m]

print q_movies.shape
```

```python
# Function that computes the weighted rating for each movie
def weighted_rating(x, m=m, C=C):
    v = x['num_voted_users']
    R = x['imdb_score']
    return (v/(v+m)*R) + (m/(v+m)*C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)

# Print the top 15 movies
q_movies[['movie_title', 'num_voted_users', 'imdb_score', 'score']].head(15)
```
