# Numpy

```python
# Convert list to numpy array
mylist = [[1, 2, 3], [4, 5, 6]]
arr = np.asarray(mylist)
print arr, type(arr)
print arr.shape
print arr.dtype
```

```python
# create a numpy array
arr1 = np.array([[4, 5], [6, 7], [8, 9]])
print arr1
print type(arr1), arr1.dtype, arr1.shape
```

```python
# create ones array
arr1 = np.ones((50,6), dtype='float64')
print arr1.shape, arr1.dtype
print arr1[36][2:]
print arr1.sum()   # total sum
print sum(arr1[0:4])[0] # row-wise sum
print sum(arr1)[3] # column wise sum
print sum(arr1).sum()
```

```python
# create Identity matrix
I = np.eye(4)
print I, I.shape
print I.dtype
```

```python
# Copy array
arr2 = arr1.copy()
print arr2.shape
print arr2[2][3]
print arr2.dtype
```

```python
# Adjusting dimensions
arr3 = arr2.reshape((10, 30))
print arr3.shape
arr4 = arr2.reshape((20,-1))
print arr4.shape
```

```python
# Random arrays
a = np.random.random((3, 4))
print a, a.shape

b = np.random.randint(4, 6, (2, 3))
print b
```

```python
# mean of array
print b.mean()
print b[0].mean()
print b[1].mean()
```

```python
# Transpose of matrix
c = b.T
print c
```

```python
# Concatenate matrices
a1 = np.zeros((2, 3))
a2 = np.ones((2, 4))
a3 = np.concatenate((a1, a2), axis=1)
print a3
print a3.shape
```

```python
# Calculate inverse of a matrix
a_inv = np.linalg.pinv(a2)
print a_inv, a_inv.shape
```

```python
# Matrix multiplication
a = np.array([[2, 3], [3, 2]])
b = np.array([[3, 5, 6], [4, 6, 2]])
c = a.dot(b)
print c, c.shape
print b.reshape(3, 2)
```

```python
# Element wise multiplication
x = np.array([[1, 1], [2, 2]])
y = np.array([[3, 3], [2, 4]])
z = x * y
print z
```

```python
# Typecasting elements of matrix
x = x.astype(float)
print x
```

```python
# Concatenate and reshape
cr = np.concatenate((a, b.reshape(-1,2)), axis=0)
print cr
print cr.shape
```

# Matplotlib

```python
# Point Plotting
plt.figure(0)
plt.plot(2, 3, 'bo')
plt.show()
```

```python
# Line plotting
x = np.linspace(2, 9, 500)
y = 2 * x

plt.figure(1)
plt.plot(x, y)
plt.show()
```

```python
y = 3 * x * x
plt.figure(2)
plt.plot(x, y)
plt.grid('on')
plt.show()
```

```python
# Plotting triangle
x = np.linspace(-2, 0, 500)
y = x + 2
plt.figure(3)
plt.title('Triangle')
plt.grid('on')
plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.plot(x, y)
plt.plot(-x, y)
plt.plot(np.linspace(-2,2, 500), np.zeros(500))
plt.show()
```

```python
# Scatter plot
x = np.random.randn(500)
y = np.random.randn(500)
plt.scatter(x, y, c='r')
plt.show()
```

```python
# Plot sine and cosine wave
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, 'r', label='sin(x)')
plt.plot(x, y2, 'b', label='cos(x)')
plt.legend(loc='best')
plt.show()
```

```python
# Subplots
x = np.linspace(2, 9, 500)
y = 2 * x
#plt.figure(1)
plt.subplot(2,2,1)
plt.plot(x, y)

y = 3 * x * x
#plt.figure(2)
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.grid('on')

x = np.linspace(-2, 0, 500)
y = x + 2
#plt.figure(3)
plt.subplot(2,2,2)
plt.title('Triangle')
plt.grid('on')
plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.plot(x, y)
plt.plot(-x, y)
plt.plot(np.linspace(-2,2, 500), np.zeros(500))

x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
#plt.figure(4)
plt.subplot(2, 2, 4)
plt.plot(x, y1, 'r', label='sin(x)')
plt.plot(x, y2, 'b', label='cos(x)')
plt.legend(loc=1)
plt.show()
```

# OpenCV

```python
# Image loading, plotting and color conversion
img = cv2.imread('/home/puja/Downloads/dog.jpg')
plt.subplot(2,2,1)
plt.imshow(img)

im = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(2,2,2)
plt.imshow(im)
print im.shape

im2 = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)
plt.subplot(2,2,3)
plt.imshow(im2, cmap='gray')
```

```python
# Resize an image
imr = cv2.resize(im, (100, 100))
print imr.shape
plt.imshow(imr)
```
